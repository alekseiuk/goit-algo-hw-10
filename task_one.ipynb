```python
import pulp
```

```python
model = pulp.LpProblem("Production_Optimization", pulp.LpMaximize)
```

```python
lemonade = pulp.LpVariable("Lemonade", lowBound=0, cat="Integer")
juice = pulp.LpVariable("Fruit_Juice", lowBound=0, cat="Integer")
```

```python
model += lemonade + juice, "Maximize_Total_Production"
```

```python
model += 2 * lemonade + 1 * juice <= 100, "Water_Constraint"
model += 1 * lemonade <= 50, "Sugar_Constraint"
model += 1 * lemonade <= 30, "Lemon_Juice_Constraint"
model += 2 * juice <= 40, "Fruit_Puree_Constraint"
```

```python
model.solve()
```

```python
print("Status:", pulp.LpStatus[model.status])
```

```python
print("Amount of lemonade:", lemonade.varValue)
```

```python
print("Amount of fruit juice:", juice.varValue)
```

```python
print("Total:", pulp.value(model.objective))
```
